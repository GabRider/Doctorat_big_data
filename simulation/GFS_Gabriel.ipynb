```python
import xarray as xr
import requests

# Define the GFS data URL (adjust based on the server or AWS source)
gfs_url = "https://nomads.ncep.noaa.gov/pub/data/nccf/com/gfs/prod/gfs.20231201/00/atmos/gfs.t00z.pgrb2.0p25.f000"

# Define the variables and pressure levels to filter
variables = ["TMP", "UGRD", "VGRD", "HGT", "SPFH"]
pressure_levels = [100, 200, 250, 500, 850, 925, 1000]

# Load the data using xarray and cfgrib
ds = xr.open_dataset(gfs_url, engine="cfgrib", filter_by_keys={"typeOfLevel": "isobaricInhPa"})

# Filter by required variables and pressure levels
filtered_ds = ds.sel(isobaricInhPa=pressure_levels)[variables]

# Save to a local NetCDF file
filtered_ds.to_netcdf("filtered_gfs_data.nc")

print("Downloaded and filtered GFS data saved as 'filtered_gfs_data.nc'")
```
